## Question 1 

Comparing $D_N^2$ and $D_N^{(2)}$

### Solution

The explicit form of $D_N$ is:
$\newcommand{\bkt}[1]{\left(#1\right)}$
$$
\begin{align*}
D_N(i,j) & = \begin{cases}
\bkt{-1}^{i-j} \dfrac12 \cot \bkt{\dfrac{(i-j)h}2} & \text{if } i \neq j\\
0 & \text{if }i=j
\end{cases}
\end{align*}
$$
The form of $D_N^{(2)}$ is:
$$
\begin{align*}
D_N^{(2)}(i,j) & = \begin{cases}
\bkt{-1}^{i-j+1} \csc^2 \bkt{\dfrac{(i-j)h}2} & \text{if } i \neq j\\
\dfrac{-\pi^2}{3h^2} - \dfrac{1}{6} & \text{if }i=j
\end{cases}
\end{align*}
$$

The matrices for $N=2$:
$$D_2 = \begin{bmatrix}
0 & \dfrac12\cot\bkt{h/2} \\
-\dfrac12\cot\bkt{h/2} & 0 \\
\end{bmatrix}
D_2^2 = \begin{bmatrix}
-\dfrac14\cot^2\bkt{h/2} & 0 \\
0 & -\dfrac14 \cot^2\bkt{h/2} \\
\end{bmatrix}
D_2^{(2)} = \begin{bmatrix}
\dfrac{-\pi^2}{3h^2} - \dfrac{1}{6} & \dfrac12 \csc^2\bkt{h/2} \\
\dfrac12 \csc^2\bkt{h/2} & \dfrac{-\pi^2}{3h^2}-\dfrac16 \\
\end{bmatrix}
$$

The matrices for $N=4$:
$$D_4 = \begin{bmatrix}
0 & \dfrac12\cot\bkt{h/2} & -\dfrac12\cot\bkt{2h/2} & \dfrac12\cot\bkt{3h/2}\\
-\dfrac12\cot\bkt{h/2} & 0 & \dfrac12\cot\bkt{h/2} & -\dfrac12\cot\bkt{2h/2} \\
\dfrac12\cot\bkt{2h/2}& -\dfrac12\cot\bkt{h/2} & 0&\dfrac12\cot\bkt{h/2}\\
-\dfrac12\cot\bkt{3h/2}& \dfrac12\cot\bkt{2h/2}&-\dfrac12\cot\bkt{h/2}&0\\
\end{bmatrix}
$$

$$
D_4^2(i,j) =
\sum_{k=1}^4 \dfrac14 (-1)^{i-j} \cot\bkt{\dfrac{(i-k)h}{2}} \cot\bkt{\dfrac{(k-j)h}{2}}  \forall i,j \& \text{ If $i,j=k$, the contribution is 0, not $\cot 0=\infty$}
$$

$$
D_4^{(2)}(i,j) =
\begin{cases}
\bkt{-1}^{i-j+1} \csc^2 \bkt{\dfrac{(i-j)h}2} & \text{if } i \neq j\\
\dfrac{-\pi^2}{3h^2} - \dfrac{1}{6} & \text{if }i=j
\end{cases}
\text{   }
=
\begin{cases}
-\dfrac14\bkt{-1}^{i-j}\left[ 2+2\cot^2 \bkt{\dfrac{(i-j)h}2} \right]& \text{if } i \neq j\\
\dfrac{-\pi^2}{3h^2} - \dfrac{1}{6} & \text{if }i=j
\end{cases}
$$

Therefore, the matrices $D_N^2$ and $D_N^{(2)}$ are different.

## Question 2

Deriving $S'_N(x_j)$ and $S''_N(x_j)$

### Solution

$$
\begin{align*}
S_N(x-x_m) &= \dfrac{\sin{\dfrac{\pi(x-x_m)}{h}}}{\dfrac{2\pi}{h} \tan\bkt{\dfrac{x-x_m}{2}}} \\ \\
S'_h(x-x_m) &= \frac{d}{dx}\dfrac{\sin{\dfrac{\pi(x-x_m)}{h}}}{\dfrac{2\pi}{h} \tan\bkt{\dfrac{x-x_m}{2}}} \\ \\
&= \dfrac{h}{2\pi}\left[\dfrac{ \dfrac{\pi}{h} \cos \dfrac{\pi(x-x_m)}{h} \tan \dfrac{(x-x_m)}{2} - \dfrac{1}{2} \sin{\dfrac{\pi(x-x_m)}{h} \sec^2 \dfrac{(x-x_m)}{2} }}{\tan^2\dfrac{(x-x_m)}{2}} \right]\\ \\
&=   \dfrac{h}{2\pi} \left[\dfrac{\pi}{h} \dfrac{ \cos \dfrac{\pi(x-x_m)}{h}}{ \tan \dfrac{(x-x_m)}{2}} - \dfrac{1}{2} \dfrac{ \sin \dfrac{\pi(x-x_m)}{h} } {\sin^2 \dfrac{(x-x_m)}{2} } \right]
\end{align*}
$$

When $x_j = jh, j \in \mathbb{Z}$, $(j-m)=k$, and $\frac{\pi}{h} = \frac{N}{2}$, we get two cases for the expression above: one where $k \equiv 0 (\mod N)$ and other when $k \not\equiv 0 (\mod N)$. Further simplification is because $\cos(n\pi) = (-1)^n$ and $\sin{n\pi} = 0$ when $n\neq0$. If $n=0$, $\sinc{n\pi}\rightarrow1$ and $\lim_{x\rightarrow0}\frac{\tan x}{x}=1$.

Therefore, 
$$
\begin{align*}
S'_N(x_j) & = \begin{cases}
\dfrac12 {(-1)}^{j}\cot\bkt{\frac{jh}{2}} & \text{if } j \not\equiv 0 (\mod N)\\
0 & \text{if }j\equiv 0(\mod N)
\end{cases}
\end{align*}
$$

On differentiating $S'_N(x-x_m)$ further, we get:

$$
\newcommand{\sinc}[1]{\mathrm{sinc}\left(#1\right)}
\begin{align*}
S''_N(x-x_m) &= \dfrac{h}{2\pi}\left\{ \left[\dfrac{-\pi^2}{h^2} \dfrac{ \sin \dfrac{\pi(x-x_m)}{h}}{\tan \dfrac{(x-x_m)}{2}} - \dfrac{\pi}{2h} \dfrac{ \cos \dfrac{\pi(x-x_m)}{h}}{\sin^2 \dfrac{(x-x_m)}{2}} \right]+ \left[  \dfrac{-\pi}{2h} \dfrac{ \cos \dfrac{\pi(x-x_m)}{h}}{\sin^2 \dfrac{(x-x_m)}{2}} + \dfrac{1}{4} \dfrac{ \sin \dfrac{\pi(x-x_m)}{h} \sin (x-x_m)}{\sin^4 \dfrac{(x-x_m)}{2}} \right] \right\} \\ \\
&= \dfrac{h}{2\pi} \left\{ \dfrac{\sin \dfrac{\pi(x-x_m)}{h}}{\tan \dfrac{(x-x_m)}{2}} \left[\dfrac{-\pi^2}{h^2} + \dfrac{1}{\sin^2 \dfrac{(x-x_m)}{2}} \right] -  \dfrac{\pi}{h}\dfrac{\cos \dfrac{\pi(x-x_m)}{h}}{\sin^2 \dfrac{(x-x_m)}{2}}\right\} \\ \\
\end{align*}
$$

When applying the limiting conditions as with the first derivative, the second derivative becomes:
$$
\begin{align*}
S''_N(x_j) & = \begin{cases}
-\dfrac12 \dfrac{(-1)^{j}}{\sin^2\bkt{\frac{jh}{2}}} & \text{if } j \not\equiv 0 (\mod N)\\
-\dfrac{\pi^2}{3h^2} -\dfrac16 & \text{if }j\equiv 0(\mod N)
\end{cases}
\end{align*}
$$

## Question 3

Deriving $D_N^{(3)}$

### Solution

There are at least two ways to obtain this matrix. One is through calculus and the other is by using the discrete fourier transform. In this solution, the calculus route is chosen. 

From the previous question, the value of $S''_N(x-x_m)$ is already known. This can be differentiated one more time. The resultant expression is:
$$
\newcommand{\pixh}[0]{\dfrac{\pi(x-x_m)}{h}}
\newcommand{\xhalf}[0]{\dfrac{(x-x_m)}{2}}
\begin{align*}
\\
S'''_N(x-x_m) &= \dfrac{h}{2\pi}
\left\{
\dfrac{-\pi^3}{h^3} \dfrac{ \cos \pixh}{\tan \xhalf}
+ \dfrac{\pi^2}{2h^2} \dfrac{ \sin \pixh }{\sin^2 \xhalf}
+ \dfrac{\pi}{2h} \dfrac{\cos \pixh }{ \tan \xhalf \sin^2 \xhalf}
- \dfrac{1}{2} \dfrac{\sin \pixh }{\sin^2 \xhalf \tan^2 \xhalf} \\
- \dfrac{1}{4} \dfrac{\sin \pixh}{\sin^4 \xhalf} 
+ \dfrac{\pi^2}{h^2} \dfrac{\sin \pixh}{\sin^2 \xhalf}
+ \dfrac{\pi}{h} \dfrac{\cos \pixh}{\sin^2 \xhalf \tan \xhalf}
\right\} \\ \\
&= \dfrac{h}{2\pi}
\left\{
\dfrac{-\pi^3}{h^3} \dfrac{ \cos \pixh}{\tan \xhalf}
+ \dfrac{3\pi^2}{2h^2} \dfrac{ \sin \pixh }{\sin^2 \xhalf}
+ \dfrac{3\pi}{2h} \dfrac{\cos \pixh }{ \tan \xhalf \sin^2 \xhalf} \\
- \dfrac{1}{2} \dfrac{\sin \pixh }{\sin^2 \xhalf}
\left[ \dfrac{1}{\tan^2 \xhalf} + \dfrac{1}{2\sin^2 \xhalf} \right]
\right\}
\end{align*}
$$

Substituting $\frac{x-x_m}{2} = \frac{kh}{2} = \frac{k\pi}{N}$, we get that
$$
\begin{align*}
S'''_N(x_j) & = \begin{cases}
-\dfrac{\pi^2}{2h^2} \dfrac{(-1)^{j}}{\tan\bkt{\frac{jh}{2}}} -\dfrac{3}{4} \dfrac{(-1)^{j}}{\sin^2\bkt{\frac{jh}{2}} \tan\bkt{\frac{jh}{2}}} & \text{if } j \not\equiv 0 (\mod N)\\
-\dfrac{3\pi^3}{2h^3} & \text{if }j\equiv 0(\mod N)
\end{cases}
\end{align*}
$$

## Question 8

Spectral Method for space, Euler Explicit for time.

### Solution

```python
# SPECTRAL WITH EULER EXPLICIT
# Grid, variable coefficient and initial data
import scipy as sp;
from scipy import linalg;
from scipy import sparse;
pi = sp.pi;
N  = 128;      # Number of grid points
h  = 2*pi/N; # Grid spacing
print(h)         # Display grid spacing
x  = h*sp.arange(1,N+1); # Grid points
vec           = sp.zeros(N);
vec[1:N]   = 0.5*(-1)**sp.arange(1,N)/sp.tan(sp.arange(1,N)*h/2);
D              = sp.linalg.circulant(vec);

tinit           = 0;         # Intial time
tmax         = 8;         # Maximum time
nsteps       = 10000;    # Number of steps
dt              = (tmax-tinit)/nsteps;  # time step
print(dt)           # Display time step
# c = 0.2; # Constant Wave speed
c = 0.2+sp.sin(x-1)**2; # Variable Wave speed
v   = sp.exp(-100*(x-1)**2); # Initial condition
# plotstep = 10; # How frequent should I plot?
# tplot      = plotstep*dt;
# nplots    = nsteps//plotstep;
data = sp.zeros((nsteps+1,N));
data[0,:] = v;
tdata = sp.zeros(nsteps+1);
tdata[0] = tinit;
for i in sp.arange(0,nsteps):
    v = v-dt*c*(D.dot(v));
    data[i+1,:] = v;
    tdata[i+1] = tdata[i]+dt;

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np


X, T = np.meshgrid(x, tdata, sparse=False, indexing='ij')

fig = plt.figure();
ax = fig.add_subplot(111, projection='3d');
# X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_wireframe(X.T, T.T, data, rstride=20, cstride=20)
ax.view_init(60,-75)
plt.show()

plt.pcolor(x, tdata, data)
plt.colorbar()
plt.title('Spectral with Euler Explicit')
plt.show()
```

The spectral derivative ensures that there are not many secondary waves. However, the explicit scheme leads to dampening of the amplitude of the wave. The computational time will be lesser than with Leap-Frog.

## Question 9

2nd Order Finite Difference with Leap Frog

### Solution

```python
# SECOND ORDER FINITE DIFFERENCE WITH LEAP FROG
# Grid, variable coefficient and initial data
import scipy as sp;
from scipy import linalg;
from scipy import sparse;
pi = sp.pi;
N  = 128;      # Number of grid points
h  = 2*pi/N; # Grid spacing
print(h)         # Display grid spacing
x  = h*sp.arange(1,N+1); # Grid points
v           = sp.zeros(N);
v[1]        = -0.5/h;
v[-1]       = 0.5/h;
D           = sparse.csr_matrix(sp.linalg.circulant(v));

tinit           = 0;         # Intial time
tmax         = 8;         # Maximum time
nsteps       = 800;    # Number of steps
dt              = (tmax-tinit)/nsteps;  # time step
print(dt)           # Display time step
# c = 0.2; # Constant Wave speed
c = 0.2+sp.sin(x-1)**2; # Variable Wave speed
v   = sp.exp(-100*(x-1)**2); # Initial condition
vold = sp.exp(-100*(x-c*dt-1)**2); # Second initial condition for leap frod
# plotstep = 10; # How frequent should I plot?
# tplot      = plotstep*dt;
# nplots    = nsteps//plotstep;
data = sp.zeros((nsteps+1,N));
data[0,:] = v;
tdata = sp.zeros(nsteps+1);
tdata[0] = tinit;
for i in sp.arange(0,nsteps):
    vnew = vold-2*dt*c*(D.dot(v));
    vold = v;
    v = vnew;
    data[i+1,:] = v;
    tdata[i+1] = tdata[i]+dt;

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np


X, T = np.meshgrid(x, tdata, sparse=False, indexing='ij')

fig = plt.figure();
ax = fig.add_subplot(111, projection='3d');
# X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_wireframe(X.T, T.T, data, rstride=20, cstride=20)
ax.view_init(60,-75)
plt.show()

plt.pcolor(x, tdata, data)
plt.colorbar()
plt.title('Second order with Leap Frog')
plt.show()
```

Because of Leap-Frog, the amplitude doesn't decrease much. However, there a lot of secondary waves because of Finite Difference (as opposed to Spectral Derivatives).
